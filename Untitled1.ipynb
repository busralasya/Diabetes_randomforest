```python
df_compare=df_raw.drop("Outcome",axis=1)
```

```python
df_compare
```

```python
y_compare=df_raw["Outcome"]
```

```python
y_compare
```

```python
def split_train_val(df_raw,n): 
    
    return df_raw[:n].copy(), df_raw[n:].copy()
```

```python
n_valid = 200  
n_train = len(df_raw)-n_valid
x_train, x_valid = split_train_val(df_compare, n_train)
y_train, y_valid = split_train_val(y_compare, n_train)

x_train.shape, y_train.shape, x_valid.shape, y_valid.shape
```

```python
def rmse(x,y): 
    return math.sqrt(((x-y)**2).mean())
```

```python
def print_score(m):
    
    print(f"RMSE of train set {rmse(m.predict(x_train), y_train)}")
    print(f"RMSE of validation set {rmse(m.predict(x_valid), y_valid)}")
    print(f"R^2 of train set {m.score(x_train, y_train)}")
    print(f"R^2 of validation set {m.score(x_valid, y_valid)}")
```

```python
m = RandomForestRegressor(n_jobs=-1)
%time m.fit(x_train, y_train)
print_score(m)
```

```python
def get_sample(df_raw,n):

    idxs = np.random.permutation(len(df_raw))[:n]
    return idxs, df_raw.iloc[idxs].copy()
```

```python
n_valid = 200 # same as Kaggle's test set size
n_train = len(df_raw)-n_valid
x_train, x_valid = split_train_val(df_compare, n_train)
y_train, y_valid = split_train_val(y_compare, n_train)

x_train.shape, y_train.shape, x_valid.shape, y_valid.shape
```

```python
idxs, x_train = get_sample(x_train, 200)
y_train = y_train[idxs]
```

```python
m = RandomForestRegressor(n_jobs=-1)
%time m.fit(x_train, y_train)
print_score(m)
```

```python
def print_score(m):
    
    print(f"RMSE of train set {rmse(m.predict(x_train), y_train)}")
    print(f"RMSE of validation set {rmse(m.predict(x_valid), y_valid)}")
    print(f"R^2 of train set {m.score(x_train, y_train)}")
    print(f"R^2 of validation set {m.score(x_valid, y_valid)}")
    if hasattr(m, "oob_score_"):
        print(f"OOB score: {m.oob_score_}")
```
